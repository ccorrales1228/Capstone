# Data Wrangling

## Exploring the Target (loan status)

```python
copy_accepted_df['loan_status'].value_counts(dropna=False)
```

Need to consider removing or combining some of variables, need to focus on our target which is deafult and non default.

At this point and purpose of this project, we will focus on charge off and fully paid.

If time permitting, I will then consider combining variables.


```python
copy_accepted_df = copy_accepted_df.loc[copy_accepted_df['loan_status'].isin(['Fully Paid', 'Charged Off'])]
```

```python
copy_accepted_df.shape
```

## Dealing with missing data

```python
#Checking missing values of the features by percentage

missing_percent = copy_accepted_df.isna().mean().sort_values(ascending = False).round(4) * 100
```

```python
missing_percent
```

```python
drop_missing = sorted(list(missing_percent[missing_percent > 30].index))
```

```python
drop_missing
```

```python
len(drop_missing)
```

```python
copy_accepted_df.drop(labels=drop_missing, axis=1, inplace=True)
```
